```python
#backtracting

"""
Exemple: problème des n reines

Soit un pateau n x n. Placer n reines d'échecs de sorte qu'aucune reine n'en menace une autre

Le principe du backtracing consiste à explorer l'arbre des possibilités et à faire demitour si on arrive à une impasse
"""
import numpy as np


def case_interdite(n, i, j, plateau):
    for k in range(n):
        plateau[k,j] = False
        plateau[i, k] = False
    Diagsup = [ (x, y) for x, y in range(n) if x - y == i - j]
    Diaginf = [ (x, y) for x, y in range(n) if x + y == i + j]
    
    plateau[Diagsup] = False
    plateau[Diaginf] = False
    return plateau




def case_interdite_plus(n, reines, plateau):
    for (i,j) in reines:
        case_interdites(n, i, j, plateau)
    return plateau



def jeux_reines(n):
    plateau = np.full((n,n), True)
    l_reines = []
    i = len(l_reines)
    while i < n:
        for j in range(n):
            plat_test = plateau.copy()
            l_reines.append(j)
            case_interdite(n, i-1,j, plat_test)
            conflit = plat_test[i+1,] == False
            if not conflit:
                i +=1
                break
            else:
                l_reines.pop()
        l_reines.pop()
    return l_reines


    
                
            
            
            


    
    
    
    

```
