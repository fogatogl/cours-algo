```python
#TP
import numpy as np
#1
"""X est une matrice (10,2) à coef aléatoires"""
X = np.random.normal(size = (10,2))

#2
""" """
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1])

#3
"""Construire une matrice D de distance entre le point i et le point j de X
Quel est le format de D?"""
D= np.zeros((10,10))

#faux
for i in range(10):
    for j in range(10): 
        D[i][j] = np.sqrt((X[i][0]-X[j][0])**2 + (X[i][1]-X[j][1])**2)
        
"""méthode par slicing"""
#calcule différence coef par 'translation'
X1 = X[:, np.newaxis, :] #(10, 1, 2)
X2 = X[np.newaxis, :, :] #(1, 10, 2)

X3 = X1 - X2 #(10, 10, 2)
X4 = X3**2
X5 = X4.sum(axis=2) #(10, 10) calcule (xi-xj)^2 + (yi-yj)^2
X6 = np.sqrt(X5) #(10, 10)


print(D)
print("####")
print(X6)

ordreX = X6.argsort()
ordreX
```
